```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import Perceptron
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('iris.data.csv', names=['Sepal length', 'Sepal width', 'Petal length','Petal width', 'Species'], index_col=False)
```

```python
df.head()
```

```python
df.shape
```

```python
sns.scatterplot(x=df['Sepal length'], y=df['Petal length'], hue=df['Species'])
```

```python
sns.scatterplot(x=df['Sepal width'], y=df['Petal width'], hue=df['Species'])
```

```python
df['Species'].value_counts()
```

```python
df['Species'] = df['Species'].map({'Iris-versicolor': 0, 'Iris-virginica': 1, 'Iris-setosa': 2})
```

```python
df.sample(5)
```

```python
X = df.iloc[:,:4]
Y = df.iloc[:,-1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state=42, test_size=0.2)

X_train.shape, X_test.shape
```

```python
clf = Perceptron()
clf.fit(X_train, y_train)
```

```python
clf.intercept_
```

```python
clf.coef_
```

```python
y_pred = clf.predict(X_test)
```

```python
print(accuracy_score(y_test, y_pred))
```
